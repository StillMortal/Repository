```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 500)

plt.rcParams.update({'font.size': 26})

general_table = pd.read_excel('C:\\Users\\Akilla2\\Desktop\\notebooks\\Dissertation\\students\\1_2_БН_БФ_БИ _2020.xlsx')

general_table['Фамилия Имя'] = general_table['Фамилия'].str.cat(general_table['Имя'], sep=' ')
general_table.dropna(subset=['Фамилия Имя'], inplace=True)
general_table = general_table[general_table['Фамилия Имя'] != 'Фамилия Имя'].reset_index(drop=True)
general_table
```

```python
students = pd.read_excel('C:\\Users\\Akilla2\\Desktop\\notebooks\\Dissertation\\students\\all_entrants.xlsx')

students = students.rename(columns={'Unnamed: 1': 'Фамилия Имя'})

students.drop(
    '30. Направление подготовки 44.03.01 Педагогическое образование, направленность (профиль) «Образование в области иностранного языка» (Волховский филиал)', 
    axis=1, 
    inplace=True
)

students
```

```python
dataframe_merge = pd.merge(general_table, students, how='left', on=['Фамилия Имя'])

dataframe_merge.rename(columns={'%': '% for the discipline', 'Unnamed: 2_y': 'Points_for_the_exam'}, inplace=True)

dataframe_merge
```

```python
students_and_their_results = dataframe_merge.filter(['Фамилия Имя', 'Points_for_the_exam', '% for the discipline'])

students_and_their_results = students_and_their_results.dropna(subset=['Points_for_the_exam']).reset_index()

students_and_their_results
```

```python
# print(len(general_table), len(students_and_their_results))

df = pd.DataFrame(
    {
        'группы': [1, 2],
        'студенты': [len(general_table), len(students_and_their_results)]
    }
)

# ax = df.hist(bins=2, alpha=0.5, color='g', orientation='horizontal')

# sns.histplot((len(general_table), len(students_and_their_results)), x='students', shrink=.8)

df.plot(kind='bar', x='группы', y='студенты', rot=0, color='black', figsize=(20, 10), grid=True)
```
